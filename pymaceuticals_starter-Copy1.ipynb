## Observations and Insights 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
merged_mice = pd.merge(mouse_metadata,study_results,on="Mouse ID",how="outer")

# Display the data table for preview
merged_mice.head()
```

```python
# Checking the number of mice.
merged_mice["Mouse ID"].nunique()
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint.   
# dupe_mice = merged_mice.loc[merged_mice.duplicated(subset=["Mouse ID","Timepoint"], keep = False)]
dupe_mice = merged_mice.loc[merged_mice.duplicated(subset=["Mouse ID","Timepoint"])]
dupe_mice["Mouse ID"].unique()
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
# df.loc[df['column_name'] == some_value]
# != does not equal

clean_mice = merged_mice.loc[merged_mice["Mouse ID"] != "g989"]
clean_mice.head()
```

```python
# Checking the number of mice in the clean DataFrame.
clean_mice["Mouse ID"].nunique()
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# (1) Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# (2) mean, median, variance, standard deviation, and SEM of the tumor volume. 
# (3) Assemble the resulting series into a single summary dataframe.

clean_mice["Drug Regimen"].unique()
grouped_mice = clean_mice.groupby("Drug Regimen")
mean_mice = grouped_mice.mean()["Tumor Volume (mm3)"]
median_mice = grouped_mice.median()["Tumor Volume (mm3)"]
var_mice = grouped_mice.var()["Tumor Volume (mm3)"]
std_mice = grouped_mice.std()["Tumor Volume (mm3)"]
SEM_mice = grouped_mice.sem()["Tumor Volume (mm3)"]

summary_mice = pd.DataFrame({"Tumor Mean":mean_mice,
                            "Tumor Median":median_mice,
                            "Tumor Variance":var_mice,
                            "Tumor Standard Deviation":std_mice,
                            "Tumor SEM":SEM_mice})

summary_mice.round(2)
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# (above)

# df.agg(['sum', 'min'])
# Using the aggregation method, produce the same summary statistics in a single line
grouped_mice.agg(["mean","median","var","std","sem"])["Tumor Volume (mm3)"].round(2)
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
# Group the clean data by mouse ID
grouped_time = clean_mice.groupby("Mouse ID")

# Run the max function, referring to timepoint
grouped_time = grouped_time.max()["Timepoint"]
grouped_time.head()

# Merge this group df with the original dataframe to get the tumor volume at the last timepoint
# Merge data frames
merged_time = pd.merge(merged_mice,grouped_time,on=["Mouse ID","Timepoint"])
merged_time.head().round(2)
```

```python
#merged_time.loc[merged_time["Drug Regimen"]=="Capomulin"]
merged_time[merged_time["Drug Regimen"]=="Capomulin"]
```

```python
# Put treatments into a list for for loop (and later for plot labels)
drugs_list = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]

# Create empty list to fill with tumor vol data (for plotting)
tumor_vol_list = []

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
# Make a loop that will go thru each object in drugs list
for drug in drugs_list:
    filtered_df = merged_time.loc[merged_time["Drug Regimen"] == drug]
    
    # Locate the rows which contain mice on each drug and get the tumor volumes
    # Add subset
    tumor_vol_list.append(filtered_df["Tumor Volume (mm3)"])  
    
    # Determine outliers using upper and lower bounds
    quantiles = filtered_df.quantile([0.25,0.5,0.75])
    quantiles = quantiles["Tumor Volume (mm3)"]
    lower_q = quantiles[.25]
    upper_q = quantiles[0.75]
    iqr = upper_q - lower_q
    lower_bound = lower_q - (1.5*iqr)
    upper_bound = upper_q + (1.5*iqr)
    outliers = filtered_df.loc[(filtered_df["Tumor Volume (mm3)"] < lower_bound) | (filtered_df["Tumor Volume (mm3)"] > upper_bound)]
    print(f"{drug}'s potential outliers: {outliers}")
```

```python
tumor_vol_list[0]
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest & format
outlier = dict(markerfacecolor="blue",markersize=10)
plt.boxplot(tumor_vol_list, labels=drugs_list,flierprops=outlier)
plt.ylabel("Final Tumor Volume (mm3)")
plt.xlabel("Drug")
plt.title("Boxplot of Tumor Volume (mm3)")
plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a mouse treated with Capomulin & format
capomulin_filter = clean_mice.loc[clean_mice["Mouse ID"] == "s185"]
plt.plot(capomulin_filter["Timepoint"],capomulin_filter["Tumor Volume (mm3)"])
plt.ylabel("Tumor Volume (mm3)")
plt.xlabel("Timepoint (s)")
plt.title("Line Plot of Timepoint (s) vs Tumor Volume (mm3)")
plt.show()
```

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen
# Find the data for only the capomulin regimen
capomulin_filter2 = clean_mice.loc[clean_mice["Drug Regimen"] == "Capomulin"]

# Group the data by mouse ID and then take the mean of all the data (to find avg tumor vol)
grouped_mice2 = capomulin_filter2.groupby("Mouse ID").mean()

# Create the scatter plot
plt.scatter(grouped_mice2["Weight (g)"],grouped_mice2["Tumor Volume (mm3)"])
plt.ylabel("Tumor Volume (mm3)")
plt.xlabel("Weight (g)")
plt.title("Scatter Plot of Weight (g) vs Tumor Volume (mm3)")
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen

# Find and print the correlation for the scatter plot data
correlation = st.pearsonr(grouped_mice2["Weight (g)"],grouped_mice2["Tumor Volume (mm3)"])
print(f"The correlation between both factors is {round(correlation[0],2)}")

# Show the linear regression model for the scatter plot data
x_values = grouped_mice2["Weight (g)"]
y_values = grouped_mice2["Tumor Volume (mm3)"]
(slope, intercept, rvalue, pvalue, stderr) = st.linregress(x_values, y_values)
regress_values = x_values * slope + intercept
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.xlabel("Weight (g)")
plt.ylabel("Tumor Volume (mm3)")
plt.title("Line Plot of Weight (g) vs Tumor Volume (mm3)")
plt.show()
```
